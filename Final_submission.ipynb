```
#import raw data
old_train = pd.read_csv('/dbfs/FileStore/wids/train.csv')
old_test = pd.read_csv('/dbfs/FileStore/wids/test.csv')
```

```
#load dataset 2
df = pd.read_csv('/dbfs/FileStore/wids/features_data_0102_imputed_star_rating_and_generous_building_id.csv')
df['year_built'] = df['year_built'].astype('int')
df.head(2)
```

```
df = df[['cooling_degree_days', 'energy_star_rating',  'fac_type_mean', 'year_factor', 'snowfall_inches', 'building_id_mean',
                       'facility_type', 'days_below_20F',  'building_id', 'weather_cluster', 'floor_area','id', 'site_eui']]


df['log_floor_area'] = np.log( df['floor_area'])
df['log_site_eui'] = np.log(df['site_eui'])

#only year 6
df = df[df['year_factor']>=6]

df = df.drop(['floor_area', 'site_eui', 'year_factor'],axis=1)

df.head(2)
```

```
#split into (new) train and test set based on id
df_train = df[df["id"].isin(old_train["id"])]
df_test = df[df["id"].isin(old_test["id"])]
```

```
encoded_train  = df_train.copy()
encoded_test  = df_test.copy()

# Encode Categorical Columns 
cat_cols = ['facility_type']
for col in cat_cols:
  le = LabelEncoder()
  encoded_train[col] = le.fit_transform(encoded_train[col])
  encoded_test[col] = le.transform(df_test[col])
```

#2 Baseline models

```
#define X and y for training the model

encoded_train_log = encoded_train.copy()
encoded_train_log = encoded_train_log[encoded_train_log['year_factor']>=6]

X= encoded_train_log.drop(['id', 'site_eui', 'log_site_eui', 'floor_area',  'year_factor'], axis=1)
y = encoded_train_log['log_site_eui']

#define the models
models=[LinearRegression( ),
        Ridge(),
        Lasso(),
        KNeighborsRegressor(),
        Ridge(),
        DecisionTreeRegressor( random_state=42),
       RandomForestRegressor( random_state=42),
       XGBRegressor( random_state=42),
        BaggingRegressor( random_state=42), 
        AdaBoostRegressor( random_state=42),
        GradientBoostingRegressor( random_state=42),
       lgb.LGBMRegressor(),
       CatBoostRegressor( random_state=42),]

#create a list of model names
names =['Linear Regression', 'Ridge Regression', 'Lasso Regression', 'KNN Regressor', 'RidgeRegression', 'Decision Tree Regressor', 'Random Forest Regressor',  
        'XGB Regressor', 'Bagging Regressor', 'AdaBoost Regressor', 'Gradient Boosting Regressor', 'LGBM Regressor',  'Catboost Regressor',]


#create empty lists for the accuracy and standard deviation
rmse=[]
std = []

#peform 10-fold cross validaton on each model and append the results to the list
for model in models:
  cv = KFold(n_splits=5, shuffle = True,  random_state=1)
  n_scores = cross_val_score(model, X, y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1, error_score='raise')
  mean_score = np.mean(n_scores)*-1
  std_score = np.std(n_scores)
  rmse.append(mean_score)
  std.append(std_score)

#create a dataframe of the model performances
d = {'Model':names,'RMSE':rmse, 'Std':std}   
score_df = pd.DataFrame(d)
score_df
```

2. Catboost
